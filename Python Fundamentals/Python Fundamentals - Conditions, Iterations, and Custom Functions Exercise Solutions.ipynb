### Exercise 1
If the age is under 10, print "You are not eligible to buy the ticket."
- If the age is under 18 or above 65, print "Ticket price is 20 dollars."
- If the age is in between 18 to 65, print "Ticket price is 30 dollars."

```python
age = 72
if age < 10:
    print("You are not eligible to buy the ticket.")
elif age < 18 or age > 65:
    print("Ticket price is 20 dollars.")
else:
    print("Ticket price is 30 dollars.")

# Run the code with different values of age to check the result.
```
